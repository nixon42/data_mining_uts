```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
#load dataset
iris = load_iris()

X = iris.data
y = iris.target
X,y
```

```python
_label = 3
total_dataset = iris.data.shape[0]
split_train_ratio = 0.8
_split_train = int(total_dataset * split_train_ratio/_label)

iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_target = target_df = pd.DataFrame(iris.target, columns=["species"])
data = pd.concat([iris_df, iris_target], axis=1)

data_train = []
data_test = []
for i in range(2):
    data_train.append(data[data["species"] == i][:_split_train])
    data_test.append(data[data["species"] == i][_split_train:])

data_train = pd.concat(data_train)
data_test = pd.concat(data_test)

data_train.shape, data_test.shape
```

```python
petal_length_min = data_train["petal length (cm)"].min()
petal_length_max = data_train["petal length (cm)"].max()

petal_width_min = data_train["petal width (cm)"].min()
petal_width_max = data_train["petal width (cm)"].max()

sepal_length_min = data_train["sepal length (cm)"].min()
sepal_length_max = data_train["sepal length (cm)"].max()

sepal_width_min = data_train["sepal width (cm)"].min()
sepal_width_max = data_train["sepal width (cm)"].max()

petal_width_min, petal_width_max
```

```python
def discretize_binning(data, num_bin=10):
    """
    Binning discretization for each feature in pandas dataframe
    :param data: pandas dataframe
    :param num_bin: number of bin
    :return: pandas dataframe with discretized data
    """
    data_disc = data.copy()
    for feature in data.columns:
        data_disc[feature] = pd.qcut(data[feature], q=num_bin, labels=False, duplicates='drop')
    return data_disc

disc_data_train = discretize_binning(data_train, num_bin=10)
disc_data_test = discretize_binning(data_test, num_bin=10)
disc_data_train, disc_data_test
```
